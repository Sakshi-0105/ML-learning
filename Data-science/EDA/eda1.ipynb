```python
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
# data analisys phase , here main aim is to understand more about data

#display all colums from a data set irespective how much datcolums it has
pd.pandas.set_option('display.max_columns',None)
```

```python
## importing data
dataset=pd.read_csv('train.csv')
dataset.shape
```

```python
# Top 5 records
dataset.head()
```

```python
## in data analysis we try to find following stuf:
# 1.missing values
# 2.all numeric variable
#3.Distribution of all numeric variable
# 3.all categorical variable
#4.cardinality of all categorical variable
#5.outlier
#6.relation between dependent and independent features(price)
```

```python
#missing values
feature_with_na=[feature for feature in dataset.columns if dataset[feature].isnull().sum()>1]
feature_with_na
```

```python
## printing percentage of mean walues
for feature in feature_with_na:
    print(feature,"  " ,np.round(dataset[feature].isnull().mean(),4)," % missing values")
```

```python
## finding relationship between missing values and  dependent fearure
for feature in feature_with_na:
    data=dataset.copy()
    data[feature]=np.where(data[feature].isnull(),1,0)
    data.groupby(feature)['SalePrice'].median().plot.bar()
    plt.title(feature)
    plt.show()
```

```python
## here clearly there is a relation between missing values and price feature
# we neend to replace these null values with something in feature engineering

#since the id feature is unique so we can drop it

print(dataset.shape,dataset['Id'].count())
```

## numerical variables

```python
# list numereical variables
numeric_feature=[feature for feature in dataset.columns if dataset[feature].dtype!='O'] # o here means object

print(numeric_feature)
print(len(numeric_feature))
dataset[numeric_feature]
```

### there is temporal variable : a variable associated with time or functions of time
there is  some dependency of salePrice on these date variables also . like the year of house created or selled

```python
## fetching the year features
year_features = [ feature for feature in numeric_feature if 'Yr' in feature or 'Year' in feature]
print(year_features)
```

```python
# exploring content of year feature
for feature in year_features:
    print("{}:".format(feature), dataset[feature].unique())
```

```python
## exploring the relationship between the sales price and year sold by ploting a graph
dataset.groupby('YrSold')['SalePrice'].median().plot()
plt.xlabel('year Sold')
plt.ylabel('median Sale Price')
plt.title('Median Sale Price by Year Sold')
```

price decreases as year increases that not true , price need to be inceresed
thus check for other year_features relation with price variable

```python
for feature in year_features:
    if feature !='YrSold':
        data=dataset.copy()
        data[feature]=data['YrSold']-data[feature]
        plt.scatter(data[feature],data["SalePrice"])
        plt.title(feature)
        plt.xlabel(feature)
        plt.ylabel("SalePrice")
        plt.show()
```

```python
# numeric variables are of 2 type 
#contiguous and descrete
```

```python
# fetching descrete variables
descrete_feature=[feature for feature in numeric_feature if len(dataset[feature].unique())<25 and feature not in year_features+['Id']]
descrete_feature
```

```python
# analysing descrete feature
for feature in descrete_feature:
    data=dataset.copy()
    data.groupby(feature)["SalePrice"].median().plot.bar()
    plt.xlabel(feature)
    plt.ylabel("Median Sale Price")
    plt.title(feature)
    plt.show()
```

```python
# continuous feature
continuous_features = [feature for feature in numeric_feature if feature not in year_features+descrete_feature+['Id']]
print('continuous features:', continuous_features)
```

```python
## analsing the continuous variable
for feature in continuous_features:
    data=dataset.copy()
    data[feature].hist(bins=25)
    plt.xlabel(feature)
    plt.ylabel('Frequency')
    plt.title(feature)
    plt.show()
```

## eda part 2


```python
# we will use logorithmic transformation
for feature in continuous_features:
    data=dataset.copy()
    if  0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data['SalePrice']=np.log(data['SalePrice'])
        plt.scatter(data[feature],data['SalePrice'])
        plt.title(feature)
        plt.xlabel(feature)
        plt.ylabel('SalePrice')
        plt.show()
```

```python
## outliers detection with the help of box plot only for contiguous feature
for feature in continuous_features:
    data=dataset.copy()
    if  0 in data[feature].unique():
        pass
    else:
        data[feature]=np.log(data[feature])
        data.boxplot(column=feature)
        plt.title(feature)
        plt.ylabel(feature)
        plt.show()
```
