```python
btc=pd.read_csv('./Datavisual/Datavisual/crypto_data/bitcoin_price.csv')
ether=pd.read_csv('./Datavisual/Datavisual/crypto_data/ethereum_price.csv')
ltc=pd.read_csv('./Datavisual/Datavisual/crypto_data/litecoin_price.csv')
monero=pd.read_csv('./Datavisual/Datavisual/crypto_data/monero_price.csv')
neo=pd.read_csv('./Datavisual/Datavisual/crypto_data/neo_price.csv')
qtum=pd.read_csv('./Datavisual/Datavisual/crypto_data/qtum_price.csv')
ripple=pd.read_csv('./Datavisual/Datavisual/crypto_data/ripple_price.csv')


```

```python
btc.columns=btc.columns.map(lambda x:str(x)+'_btc')
ether.columns=ether.columns.map(lambda x:str(x)+'_ether')
ltc.columns=ltc.columns.map(lambda x:str(x)+'_ltc')
monero.columns=monero.columns.map(lambda x:str(x)+'_monero')
neo.columns=neo.columns.map(lambda x:str(x)+'_neo')
qtum.columns=qtum.columns.map(lambda x:str(x)+'_qtum')
ripple.columns=ripple.columns.map(lambda x:str(x)+'_ripple')
```

```python
btc.head()
## merging all the data
m1=pd.merge(btc,ether,how='inner',left_on='Date_btc',right_on='Date_ether')
m2=pd.merge(m1,ltc,how='inner',left_on='Date_btc',right_on='Date_ltc')
m3=pd.merge(m2,monero,how='inner',left_on='Date_btc',right_on='Date_monero')
m4=pd.merge(m3,neo,how='inner',left_on='Date_btc',right_on='Date_neo')
m5=pd.merge(m4,qtum,how='inner',left_on='Date_btc',right_on='Date_qtum')
crypto=pd.merge(m5,ripple,how='inner',left_on='Date_btc',right_on='Date_ripple')

crypto.head()




```

```python
curr=crypto[['Close_btc','Close_ether','Close_ltc','Close_monero','Close_neo','Close_ripple','Close_qtum']]
curr.head()
```

```python
## pairPlot
## plot graph for every pair of column
sns.pairplot(curr)
```

```python
## coreleation between variable 
corr=curr.corr()
corr
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(corr,cmap='YlGnBu',annot=True)
```

```python
## case study
market_df=pd.read_csv('./Datavisual/Datavisual/global_sales_data/market_fact.csv')
product_df=pd.read_csv('./Datavisual/Datavisual/global_sales_data/prod_dimen.csv')


product_df.head()
df=pd.merge(market_df,product_df,how='inner',on='Prod_id')
df.head()

# sns.boxplot(x="Product_Category",y="Profit",data=df)
```

```python
sns.boxplot(x="Product_Category",y="Sales",data=df)
plt.yscale('log')
```

```python
customer_df=pd.read_csv('./Datavisual/Datavisual/global_sales_data/cust_dimen.csv')
df=pd.merge(df,customer_df,how='inner',on='Cust_id')
df.head()
```

```python
df=df[(df.Profit<1000) & (df.Profit>-1000)]
sns.boxplot(x='Customer_Segment',y='Profit',data=df)
# plt.yscale('log')
```

```python
# set figure Size for large data set
plt.figure(num=None,figsize=(12,8),dpi=80,facecolor='w',edgecolor='k')
# specifying the hue='category variable
sns.boxplot(x='Customer_Segment',y='Profit',hue='Product_Category',data=df)
```
