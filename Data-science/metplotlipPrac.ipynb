```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ploting 2 1-D arrays

x=np.linspace(5,100,100)
y=np.linspace(10,1000,100)
plt.plot(x,y,'r^',x,y*2,'b+') # the extra character like b+ where b represent color and the symbol represents the sign used for 
plt.xlabel('X-axis') # for labelling purpose
plt.ylabel('Y-axis')
```

```python
# differnt ploting
# it also work with list as it internally convert the list into the array

plt.plot([1,4,6,8],[3,8,3,5])
plt.show()
```

```python
# axis labl and tittle
plt.plot([1,4,6,8],[3,8,3,5])
plt.xlabel("x-axix")
plt.ylabel("y-axis")
plt.title("x-y graph")
# ranges on x and y axis with min and max
# plt.xlim(4,8)
# plt.ylim(8,5)
```

```python
## figures and subplot
x= np.linspace(1,10,100)
y=np.log(x)
 # iniciate new figure explicitly
plt.figure(1)
plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("log(x)")

plt.subplot(1,2,2)
plt.plot(x,y**2)
plt.title("y=log(x)**2")

```

```python
## 4 graps
x=np.linspace(1,100,100)
plt.figure(1)
plt.subplot(2,2,1)
plt.plot(x,x)
plt.title("lineear")

plt.subplot(2,2,2)
plt.plot(x,x**2)
plt.title("square")


plt.figure(2)
plt.subplot(2,2,1)
plt.plot(x,x**3)
plt.title("cube")

plt.subplot(2,2,2)
plt.plot(x,x**4)
plt.title("four")
```
