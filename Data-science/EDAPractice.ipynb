```python
ratings=final_df.groupby(['Aggregate rating','Rating color', 'Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
```

```python
ratings
```

observation
1.when ever the rating is between 4.5 to 4.9 they are excellent
2.when ever the rating is between 4.0 to 4.4 they are very good
3.when ever the rating is between 3.5 to 3.9 they are good
4.when ever the rating is between 3.0 to 3.4 they are average
5. when the ratings is between 2.5 to 2.9  are good
6.when ever the rating is between 2.0 to 2.4 they are poor
7.when ever the rating is between 1.5 to 1.9 they are very poor
8.when ever the rating is between 1.0 to 1.4 they are bad

```python
from matplotlib import rcParams
rcParams['figure.figsize']=(12,6)
sns.barplot(x='Aggregate rating',y='Rating Count',data=ratings)
```
