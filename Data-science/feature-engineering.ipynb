```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#import tain and test data
test_df=pd.read_csv('test.csv')
train_df=pd.read_csv('train.csv')


```

```python
## merge both data
df =pd.concat([train_df,test_df])
```

```python
df.info()
df.isnull().sum()
df.drop("User_ID",axis=1,inplace=True)
```

```python
# handling categorical feature gender
pd.get_dummies(df["Gender"])
df["Gender"]=df["Gender"].map({'M':1,'F':0})
```

```python
# handling categorical feature age
pd.get_dummies(df["Age"])
df["Age"]=df.Age.map({'0-17':1, '55+':7, '26-35':3, '46-50':5, '51-55':6, '36-45':4, '18-25':2})
```

```python
dummy_city=pd.get_dummies(df["City_Category"],drop_first=True)
df=pd.concat([df,dummy_city],axis=1)
```

```python
# drop  city_category
df.drop('City_Category',axis=1)
```

```python
## missing value
print(df.isnull().sum())
df["Product_Category_1"].unique()
```

```python
# fill null values with mode 
df["Product_Category_1"]=df["Product_Category_1"].fillna(df["Product_Category_1"].mode()[0])
```
